## Importation des bibliothèques et saisie des données

```python
import numpy as np
import matplotlib.pyplot as plt

# Noms des 20 villes (individus)
villes = [f"V{i}" for i in range(1, 21)]

# Noms des 6 sports (variables)
sports = ["H.Ball", "B.Ball", "Tennis", "Gym", "Natation", "F.Ball"]

# Matrice des données brutes (20 individus x 6 variables)
X = np.array([
    [1881.9, 96.8,  14.2, 25.2, 1135.5, 278.3],
    [3369.8, 96.8,  10.8, 51.6, 1331.7, 284.0],
    [4467.4, 138.2,  9.5, 34.2, 2346.1, 312.3],
    [1862.1, 83.2,   8.8, 27.6,  972.6, 203.4],
    [3499.8, 287.0, 11.5, 49.4, 2139.4, 358.0],
    [3903.2, 170.7,  6.3, 42.0, 1935.2, 292.9],
    [2620.7, 129.5,  4.2, 16.8, 1346.0, 131.8],
    [3678.4, 157.0,  6.0, 24.9, 1682.6, 194.2],
    [3840.5, 187.9, 10.2, 39.6, 1859.9, 449.1],
    [2170.2, 140.5, 11.7, 31.1, 1351.1, 256.5],
    [3920.4, 128.0,  7.2, 25.5, 1911.5,  64.1],
    [2599.6,  39.6,  5.5, 19.4, 1050.8, 172.5],
    [2828.5, 211.3,  9.9, 21.8, 1085.0, 209.0],
    [2498.7, 123.2,  7.4, 26.5, 1086.2, 153.5],
    [2685.1,  41.2,  2.3, 10.6,  812.5,  89.8],
    [2739.3, 100.7,  6.6, 22.0, 1270.4, 180.5],
    [1662.1,  81.1, 10.1, 19.1,  872.2, 123.3],
    [2469.9, 142.9, 15.5, 30.9, 1165.5, 335.5],
    [2350.7,  38.7,  2.4, 13.5, 1253.1, 170.0],
    [3177.7, 292.1,  8.0, 34.8, 1400.0, 358.9]
])

m, n = X.shape   # m = nombre d'individus (20), n = nombre de variables (6)
print(f"Dimensions de X : {m} individus x {n} variables")
```

---
## Question 1 – Matrice X et sa transposée Xᵗ

```python
# La transposée échange lignes et colonnes : Xt est de dimension (6 x 20)
Xt = X.T

print("Matrice X (20 x 6) :")
print(X)

print("\nMatrice Xt (6 x 20) :")
print(Xt)
```

**Commentaire :** X est la matrice principale où chaque ligne = une ville et chaque colonne = un sport. La transposée Xᵗ inverse ce rôle : chaque ligne représente un sport et chaque colonne une ville.

---
## Question 2 – Liste des individus

```python
# Chaque individu est une ville avec ses 6 taux de pratique sportive
print("Liste des individus (villes) :")
for i in range(m):
    print(f"{villes[i]} : {X[i, :]}")
```

**Commentaire :** On accède à chaque ligne de X pour lire les données d'une ville. Les valeurs représentent les taux de pratique pour chaque sport (sur 100 000 jeunes).

---
## Question 3 – Extraction des variables (colonnes)

```python
# Chaque colonne de X correspond à un sport (variable statistique)
print("Variables (colonnes de X) :")
for j in range(n):
    print(f"{sports[j]} : {X[:, j]}")
```

**Commentaire :** On extrait chaque colonne sous forme de vecteur 1D de taille 20, représentant les taux de pratique de ce sport dans toutes les villes.

---
## Question 4 – Accès aux individus V3, V11, V15, V19

```python
# Les indices Python commencent à 0, donc V3 = index 2, V11 = index 10, etc.
indices = [2, 10, 14, 18]

print("Individus sélectionnés :")
for i in indices:
    print(f"{villes[i]} : {X[i]}")
```

**Commentaire :** On sélectionne 4 villes spécifiques par leurs indices. Ces villes ont des profils très différents — V3 a les taux les plus élevés en H.Ball et Natation, tandis que V15 et V19 ont des taux globalement plus faibles.

---
## Question 5 – Distances euclidiennes entre individus sélectionnés

```python
# La distance euclidienne mesure la "proximité" entre deux villes
# dans l'espace à 6 dimensions des sports
selection = X[indices]  # sous-matrice des 4 villes
noms_sel = [villes[i] for i in indices]

print("Matrice des distances euclidiennes :")
print(f"{'':8}", end="")
for nom in noms_sel:
    print(f"{nom:>12}", end="")
print()

for i in range(len(indices)):
    print(f"{noms_sel[i]:8}", end="")
    for j in range(len(indices)):
        d = np.linalg.norm(selection[i] - selection[j])
        print(f"{d:12.4f}", end="")
    print()
```

**Commentaire :** La diagonale vaut toujours 0 (distance d'une ville à elle-même). Plus la distance est grande, plus les profils sportifs des deux villes sont différents. V3 est probablement la plus éloignée des autres (taux très élevés), tandis que V15 et V19 sont plus proches entre elles (taux faibles similaires).

---
## Question 6 – Tableau X(j) : Moyenne, Variance, Écart-type

```python
# Pour chaque variable (sport), on calcule les 3 statistiques descriptives
# np.var et np.std utilisent par défaut la variance de population (diviseur m)
print(f"{'Variable':<12} {'Moyenne':>12} {'Variance':>14} {'Ecart-type':>12}")
print("-" * 52)

for j in range(n):
    moy = round(np.mean(X[:, j]), 4)
    var = round(np.var(X[:, j]), 4)
    std = round(np.std(X[:, j]), 4)
    print(f"{sports[j]:<12} {moy:>12.4f} {var:>14.4f} {std:>12.4f}")
```

**Commentaire :** H.Ball a la moyenne et la variance les plus élevées — c'est le sport le plus pratiqué mais aussi celui dont la pratique varie le plus entre les villes. Tennis et Gym ont des taux très faibles et peu variables. La Natation affiche également une forte dispersion.

---
## Question 7 – Individu moyen

```python
# L'individu moyen est le vecteur des moyennes de chaque variable
# Il représente la "ville fictive typique"
ind_moy = np.mean(X, axis=0)

print("Individu moyen (vecteur des moyennes) :")
for j in range(n):
    print(f"  {sports[j]} : {ind_moy[j]:.4f}")
```

**Commentaire :** L'individu moyen est le "centre de gravité" du nuage de points. Il sert de référence pour calculer les écarts (matrice centrée) et mesurer les variations autour de la moyenne.

---
## Question 8 – Matrice centrée Y

```python
# On soustrait l'individu moyen de chaque ligne : Y = X - moyenne
# Cela "centre" les données autour de 0 pour chaque variable
Y = X - ind_moy

print("Matrice centrée Y :")
print(np.round(Y, 4))

# Vérification : la moyenne de chaque colonne de Y doit être ≈ 0
print("\nVérification (moyenne de chaque colonne de Y, doit être ≈ 0) :")
print(np.round(np.mean(Y, axis=0), 10))
```

**Commentaire :** La matrice centrée Y contient les écarts de chaque ville par rapport à la moyenne. Les valeurs positives indiquent une pratique supérieure à la moyenne, les valeurs négatives une pratique inférieure. La somme (et donc la moyenne) de chaque colonne est bien 0.

---
## Question 9 – Fonction de calcul des variances

```python
# Fonction personnalisée qui calcule la variance de chaque variable
def variance_variables(M):
    """
    Calcule la variance de chaque colonne (variable) de la matrice M.
    Utilise la variance de population (diviseur m, pas m-1).
    Retourne un vecteur de taille n.
    """
    return np.round(np.var(M, axis=0), 4)

variances = variance_variables(X)

print("Variances des 6 variables :")
for j in range(n):
    print(f"  {sports[j]} : {variances[j]:.4f}")
```

**Commentaire :** La fonction encapsule le calcul pour pouvoir être réutilisée sur n'importe quelle matrice. H.Ball a la plus grande variance (forte dispersion entre villes), Tennis la plus faible (pratique quasi uniforme).

---
## Question 10 – Matrice de covariance V

```python
# Formule : V = (1/m) * Yᵗ · Y
# V est une matrice symétrique (6 x 6)
# V[i,j] = covariance entre la variable i et la variable j
# V[i,i] = variance de la variable i
V = (1 / m) * Y.T @ Y

print("Matrice de covariance V :")
print(np.round(V, 4))
```

---
## Question 11 – Commentaire de la matrice V

```python
print("=== Analyse de la matrice de covariance V ===")
print("\n• Éléments diagonaux (variances) :")
for j in range(n):
    print(f"  Var({sports[j]}) = {V[j,j]:.4f}")

print("\n• Plus grande covariance positive (forte liaison) :")
max_cov = 0
for i in range(n):
    for j in range(i+1, n):
        if abs(V[i,j]) > abs(max_cov):
            max_cov = V[i,j]
            paire = (sports[i], sports[j])
print(f"  Cov({paire[0]}, {paire[1]}) = {max_cov:.4f}")
```

**Commentaire :** 
- La **diagonale** de V contient les variances de chaque variable (identiques à celles calculées en Q9).
- Les **éléments hors-diagonale** sont les covariances entre paires de variables.
- Une covariance **positive** indique que les deux sports évoluent dans le même sens (quand l'un est élevé, l'autre l'est aussi).
- Une covariance **négative** indique une évolution opposée.
- La matrice V est **symétrique** : Cov(Xᵢ, Xⱼ) = Cov(Xⱼ, Xᵢ).
- H.Ball et Natation montrent vraisemblablement la plus forte covariance (deux sports très pratiqués qui varient ensemble).

---
## Question 12 – Matrice de corrélation R

```python
# On standardise Y : Z = Y / écart-type (centré-réduit)
# Formule : R = (1/m) * Zᵗ · Z
# Les coefficients de corrélation sont dans [-1, 1]
ecart_type = np.std(X, axis=0)    # vecteur des écarts-types
Z = Y / ecart_type                 # matrice des données centrées-réduites
R = (1 / m) * Z.T @ Z

print("Matrice de corrélation R :")
# Affichage avec noms de colonnes
print(f"{'':10}", end="")
for s in sports:
    print(f"{s:>9}", end="")
print()
for i in range(n):
    print(f"{sports[i]:10}", end="")
    for j in range(n):
        print(f"{R[i,j]:9.4f}", end="")
    print()
```

---
## Question 13 – Commentaire de la matrice R

```python
print("=== Analyse de la matrice de corrélation R ===")
print("\n• La diagonale vaut toujours 1 (corrélation d'une variable avec elle-même).")
print("• Corrélations fortes (|r| > 0.7) :")
for i in range(n):
    for j in range(i+1, n):
        if abs(R[i,j]) > 0.7:
            print(f"  r({sports[i]}, {sports[j]}) = {R[i,j]:.4f}  → {'positive' if R[i,j]>0 else 'négative'}")

print("\n• Corrélations faibles (|r| < 0.3) :")
for i in range(n):
    for j in range(i+1, n):
        if abs(R[i,j]) < 0.3:
            print(f"  r({sports[i]}, {sports[j]}) = {R[i,j]:.4f}")
```

**Commentaire :**
- Les coefficients de R sont normalisés entre **-1 et +1**, ce qui permet de comparer les liaisons indépendamment des unités.
- **r ≈ +1** : forte corrélation positive → les deux sports sont très liés (villes qui pratiquent l'un pratiquent l'autre).
- **r ≈ -1** : forte corrélation négative → relation inverse.
- **r ≈ 0** : pas de liaison linéaire détectable.
- H.Ball et Natation sont typiquement fortement corrélés (sports collectifs populaires dans les mêmes villes).
- Tennis et Gym ont souvent de faibles corrélations avec les autres sports (pratiques plus marginales et indépendantes).

## Question 14 – Nuages de points (représentation graphique)

```python
# On représente les 20 villes dans 3 plans factoriels différents :
# (X1, X4) = H.Ball vs Gym
# (X2, X5) = B.Ball vs Natation
# (X3, X6) = Tennis vs F.Ball

couples = [(0, 3), (1, 4), (2, 5)]  # indices des paires de variables

fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle("Nuages de points – Couples de variables", fontsize=13, fontweight='bold')

couleurs = ['steelblue', 'tomato', 'seagreen']

for ax, (i, j), couleur in zip(axes, couples, couleurs):
    ax.scatter(X[:, i], X[:, j], color=couleur, s=60, zorder=3)
    # Annoter chaque point avec le nom de la ville
    for k in range(m):
        ax.annotate(villes[k], (X[k, i], X[k, j]),
                    textcoords="offset points", xytext=(5, 3), fontsize=7)
    ax.set_xlabel(sports[i], fontsize=11)
    ax.set_ylabel(sports[j], fontsize=11)
    ax.set_title(f"({sports[i]}, {sports[j]})", fontsize=11)
    ax.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

**Commentaire des graphes :**

**Graphe (H.Ball, Gym) :**  
Le nuage montre une dispersion assez linéaire — les villes avec un fort taux de Hand-ball tendent aussi à avoir un taux de Gym plus élevé. V3 se démarque nettement comme valeur extrême (très haut en H.Ball).

**Graphe (B.Ball, Natation) :**  
La dispersion est plus large et moins structurée. Certaines villes pratiquent beaucoup la Natation sans pour autant pratiquer le Basket, suggérant une faible corrélation entre ces deux sports.

**Graphe (Tennis, F.Ball) :**  
Les deux variables ont des échelles très différentes (Tennis : 2–16, F.Ball : 64–449), ce qui produit un nuage étalé verticalement. La pratique du Tennis semble indépendante de celle du Football.
